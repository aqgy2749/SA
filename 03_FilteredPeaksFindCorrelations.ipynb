```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import glob
import os
```

```python
df_K_RMS=pd.read_csv("/p1/data/jtian/SA/output/03_FilteredPeaksFindCorrelations/FilteredPeaks3.csv")
print(df_K_RMS.head(10))
```

```python
#使用sklearn计算R²
def calculate_r2_sklearn(df, x_values=[0, 15, 30, 45, 60]):
    # 将x_values转换为二维数组
    X = np.array(x_values).reshape(-1, 1)
    r2_scores = []
    
    for idx, row in df.iterrows():
        y_values = row[1:].values
        
        # 创建线性回归模型
        model = LinearRegression()
        model.fit(X, y_values)
        
        # 预测并计算R²
        y_pred = model.predict(X)
        r2 = r2_score(y_values, y_pred)
        r2_scores.append(r2)
    
    return r2_scores
```

```python
r2_values_RMS = calculate_r2_sklearn(df_K_RMS)
df_K_RMS['R2'] = r2_values_RMS
```

```python
print(df_K_RMS.head(10))
```

```python
print(len(df_K_RMS[df_K_RMS.R2>0.9]))
print(len(df_K_RMS[df_K_RMS.R2>0.95]))
```
