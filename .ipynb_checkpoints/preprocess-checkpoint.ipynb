```python
import pandas as pd
import numpy as np
```

```python
##########################比对1中5并去除5
```

```python
#判断最小位数

# 读取数据
df = pd.read_csv("/p1/data/jtian/SA/shaixuanhou_0105.csv", header=None, names=['matrix', 'is1'])

# 判断matrix列的最小小数位数
matrix_min_decimals = float('inf')
for value in df['matrix']:
    if pd.notna(value):
        str_value = str(value)
        if '.' in str_value:
            decimal_places = len(str_value.split('.')[1])
            matrix_min_decimals = min(matrix_min_decimals, decimal_places)
        else:
            matrix_min_decimals = min(matrix_min_decimals, 0)

# 判断is1列的最小小数位数
is1_min_decimals = float('inf')
for value in df['is1']:
    if pd.notna(value):
        str_value = str(value)
        if '.' in str_value:
            decimal_places = len(str_value.split('.')[1])
            is1_min_decimals = min(is1_min_decimals, decimal_places)
        else:
            is1_min_decimals = min(is1_min_decimals, 0)

# 处理无穷大的情况（如果没有有效数据）
matrix_min_decimals = matrix_min_decimals if matrix_min_decimals != float('inf') else 0
is1_min_decimals = is1_min_decimals if is1_min_decimals != float('inf') else 0

print(f"matrix列最小小数位数: {matrix_min_decimals}")
print(f"is1列最小小数位数: {is1_min_decimals}")
print(f"两列中最小的小数位数: {min(matrix_min_decimals, is1_min_decimals)}")
```

```python


def tolerant_remove_duplicates(matrix_col, is1_col, tolerance=0.0001):
    """
    使用容差匹配从is1列中剔除与matrix列重合的元素
    
    参数:
    matrix_col: matrix列数据
    is1_col: is1列数据  
    tolerance: 容差范围，默认1e-10
    
    返回:
    不重合的is1元素列表
    """
    # 转换为numpy数组以便处理
    matrix_values = np.array(matrix_col)
    is1_values = np.array(is1_col)
    
    # 存储不重合的元素
    unique_elements = []
    
    for is1_val in is1_values:
        # 检查当前is1值是否与任何matrix值在容差范围内重合
        is_duplicate = False
        for matrix_val in matrix_values:
            if abs(is1_val - matrix_val) <= tolerance:
                is_duplicate = True
                break
        
        # 如果不重合，则保留
        if not is_duplicate:
            unique_elements.append(is1_val)
    
    return unique_elements


# 应用容差匹配剔除重合元素
unique_is1 = tolerant_remove_duplicates(df['matrix'], df['is1'])

# 创建结果DataFrame并保存
result_df = pd.DataFrame(unique_is1)
result_df.to_csv('dematrix-is1.csv', header=False, index=False)

print(f"处理完成！原始is1列有 {len(df['is1'])} 个元素")

print(f"剔除重合元素后剩下 {len(unique_is1)} 个元素")
print(f"结果已保存到 dematrix-is1.csv")
```
