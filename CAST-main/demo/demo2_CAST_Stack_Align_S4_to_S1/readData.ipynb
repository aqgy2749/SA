```python
import torch
import numpy as np

path = "/p1/data/jtian/SA/05_CAST/output_old/15_align_to_0_coords_final.data"
coords = torch.load(path, map_location="cpu")

first_key = list(coords.keys())[0]
A = coords[first_key].cpu().numpy()
B = coords["0"]
# 把前30个点转成可哈希的 tuple
# A30 = [tuple(x) for x in A[:30]]
# B30 = [tuple(x) for x in B[:30]]

# setA = set(A30)
# setB = set(B30)

# overlap = sorted(list(setA & setB))
# print("exact overlap count:", len(overlap))
# print("exact overlap points:")
# for p in overlap:
#     print(p)
import numpy as np
from sklearn.neighbors import NearestNeighbors

tol = 6 # 你可以改大一点，比如 0.5 / 1 / 2（看你坐标尺度）
A30 = A
B30 = B

nn = NearestNeighbors(n_neighbors=1).fit(B30)
dist, idx = nn.kneighbors(A30)
dist = dist[:,0]
idx = idx[:,0]

mask = dist <= tol
print("approx overlap count:", int(mask.sum()), " / ", len(A30), " (tol=", tol, ")")

print("\nMatched pairs (A_index, B_index, dist, A_point, B_point):")
for i in np.where(mask)[0]:
    j = idx[i]
    print(i, j, dist[i], A30[i], B30[j])
```
